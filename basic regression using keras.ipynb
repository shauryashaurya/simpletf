```python
# let's take the same linear regression example but do it in keras instead
# Keras is just an API spec and can work with multiple ML engines - it's default is now TensorFlow

# You'll see that working in Keras much closer to the
# identify data>define model>train>predict workflow that we are familiar with
# internally tf translates the keras API to dataflow
```

```python
# CUDA is improperly installed - so disable the use.
import os

os.environ["CUDA_VISIBLE_DEVICES"] = "-1"
```

```python
# all keras arrays are numpy arrays
import numpy as np

#
import tensorflow as tf
from tensorflow import keras
```

```python
# data
# as before
# Linear Equation: y = Wx + b
# Let's try y = 2x-1
x_train = np.array([-1.0, 0.0, 1.0, 2.0, 3.0, 4.0])
y_train = np.array([-3.0, -1.0, 1.0, 3.0, 5.0, 7.0])
```

```python
# define the model
model = tf.keras.Sequential([tf.keras.layers.Dense(units=1, input_shape=[1])])
# only one 'default' layer, with only one neuron
```

```python
# compile the model - this is an extra step in Keras - define the optimizer and loss function
model.compile(optimizer="sgd", loss="mean_squared_error")
```

```python
# fit to data
# also FORCE use of CPU as my GPU conflict is a PITA
with tf.device("/cpu:0"):
    model.fit(x_train, y_train, epochs=1500)
# from tensorflow.python.client import device_lib
# print(device_lib.list_local_devices())
```

```python
# predictions
print("predictions: ", model.predict([10.0, 20.0, 30.0, 40.0, 50.0]))
```
